#Answer
Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data. It helps you determine whether there's enough evidence to support a specific claim or hypothesis about a population parameter (like a mean or proportion).

```python
#Question 02. What is the null hypothesis, and how does it differ from the alternative hypothesis?
```

#Answer
Null Hypothesis vs. Alternative Hypothesis
In hypothesis testing, the null hypothesis and the alternative hypothesis are two competing statements about a population. Here's how they differ:

1. Null Hypothesis (H‚ÇÄ)
Definition: The null hypothesis is the default assumption that there is no effect, no difference, or no change.

Purpose: It serves as the starting point for the test ‚Äî something to try to disprove or reject.

Format: Usually includes equality (e.g., =, ‚â•, ‚â§).

Example:
Suppose you're testing a new drug:
H‚ÇÄ: The new drug has no effect on blood pressure.
(Or more specifically: Mean blood pressure with the new drug = mean blood pressure with the current drug.)

2. Alternative Hypothesis (H‚ÇÅ or Ha)
Definition: The alternative hypothesis is what you want to prove ‚Äî that there is an effect, a difference, or a change.

Purpose: It's the claim supported if you reject the null hypothesis.

Format: Usually includes inequality (‚â†, <, >).

Example:
H‚ÇÅ: The new drug does reduce blood pressure more than the current drug.
(Or: Mean blood pressure with the new drug < mean with the current drug.)



```python
#Question03.What is the significance level in hypothesis testing, and why is it important*
```

#Answer
The significance level, often denoted by the Greek letter Œ± (alpha), is the threshold used in hypothesis testing to decide whether to reject the null hypothesis.
The significance level (Œ±) is the probability of making a Type I error ‚Äî that is, rejecting the null hypothesis when it is actually true.

Common Values:
0.05 (5%) ‚Üí Most common
0.01 (1%)
0.10 (10%)
These values represent your tolerance for error. For example:
Œ± = 0.05 means you're willing to accept a 5% chance of wrongly rejecting a true null hypothesis.

Is It Important because:
1.Decision-Making Rule
The significance level sets the cutoff point for the p-value. If the p-value ‚â§ Œ±, you reject the null hypothesis.
2.Controls Risk
It controls the likelihood of making a false positive conclusion (Type I error).
3.Balances Precision and Risk
Lower Œ± (e.g., 0.01) means you're being more cautious ‚Äî reducing false positives, but increasing the risk of Type II errors (failing to detect a real effect).

#Answer
In hypothesis testing, the p-value helps you decide whether your data provides enough evidence to reject the null hypothesis.
Definition:

#Question01.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

```python
#Answer
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_data = np.array(sample_data)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    
    
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    
    
    p_value = 2 * (1 - norm.cdf(abs(z)))
    
    
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    
   
    if p_value < alpha:
        print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print("Result: Fail to reject the null hypothesis. No significant difference from the population mean.")


sample = [102, 100, 98, 105, 101, 99, 97, 103, 100, 102]
population_mean = 100
population_std = 3  

z_test(sample, population_mean, population_std)
```

Question03.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean?

```python
#Answer
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    
   
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    
   
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    
    
    print(f"Sample Mean       : {sample_mean:.4f}")
    print(f"Z-Score           : {z_score:.4f}")
    print(f"P-Value           : {p_value:.4f}")
    
    
    if p_value < alpha:
        print("Result: Reject the null hypothesis.")
    else:
        print("Result: Fail to reject the null hypothesis.")


np.random.seed(0)
sample_data = np.random.normal(loc=103, scale=5, size=40)  
population_mean = 100
population_std = 5  

one_sample_z_test(sample_data, population_mean, population_std)
```

Question04. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?

```python
#Answer
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    
    
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    
    
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    
    z_critical_low = norm.ppf(alpha / 2)
    z_critical_high = norm.ppf(1 - alpha / 2)
    
    
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')
    
    
    plt.fill_between(x, 0, y, where=(x <= z_critical_low), color='red', alpha=0.3, label='Rejection Region (Left)')
    plt.fill_between(x, 0, y, where=(x >= z_critical_high), color='red', alpha=0.3, label='Rejection Region (Right)')
    
    
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Observed Z-score = {z_score:.2f}')
    
    
    plt.title('Two-Tailed Z-Test: Decision Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    
    plt.grid(True)
    plt.show()
    
    
    if p_value < alpha:
        print("Decision: Reject the null hypothesis.")
    else:
        print("Decision: Fail to reject the null hypothesis.")


np.random.seed(123)
sample = np.random.normal(loc=102, scale=4, size=50)  
pop_mean = 100
pop_std = 5  

two_tailed_z_test(sample, pop_mean, pop_std)
```

Question05.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
   
    se = sigma / np.sqrt(n)
    
    
    z_crit_low = norm.ppf(alpha / 2)
    z_crit_high = norm.ppf(1 - alpha / 2)
    
    
    crit_val_low = mu0 + z_crit_low * se
    crit_val_high = mu0 + z_crit_high * se
    
    
    x_min = min(mu0, mu1) - 4*se
    x_max = max(mu0, mu1) + 4*se
    x = np.linspace(x_min, x_max, 1000)
    
   
    h0_dist = norm.pdf(x, mu0, se)
    h1_dist = norm.pdf(x, mu1, se)
    
    plt.figure(figsize=(12, 7))
    
    
    plt.plot(x, h0_dist, label='H0: Null Hypothesis Distribution', color='blue')
    plt.plot(x, h1_dist, label='H1: Alternative Hypothesis Distribution', color='green')
    
    
    plt.fill_between(x, 0, h0_dist, where=(x <= crit_val_low), color='red', alpha=0.3, label='Type I Error (Œ±)')
    plt.fill_between(x, 0, h0_dist, where=(x >= crit_val_high), color='red', alpha=0.3)
    
    
    plt.fill_between(x, 0, h1_dist, where=( (x > crit_val_low) & (x < crit_val_high) ), color='orange', alpha=0.4, label='Type II Error (Œ≤)')
    
    
    plt.axvline(crit_val_low, color='red', linestyle='--')
    plt.axvline(crit_val_high, color='red', linestyle='--')
    
    
    plt.title('Type I (Œ±) and Type II (Œ≤) Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    
    plt.show()
    
    
    beta = norm.cdf(crit_val_high, mu1, se) - norm.cdf(crit_val_low, mu1, se)
    
    print(f"Significance level (Type I error, Œ±): {alpha}")
    print(f"Type II error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.4f}")


mu0 = 100      
mu1 = 105       
sigma = 15     
n = 30          
alpha = 0.05    

plot_type1_type2_errors(mu0, mu1, sigma, n, alpha)

```

Question06. Write a Python program to perform an independent T-test and interpret the results?

```python
#Answer
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    
    mean1 = np.mean(group1)
    mean2 = np.mean(group2)
    
    
    t_stat, p_value = ttest_ind(group1, group2, equal_var=True)  
    
    
    print(f"Group 1 Mean: {mean1:.4f}")
    print(f"Group 2 Mean: {mean2:.4f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
   
    if p_value < alpha:
        print("Result: Reject the null hypothesis. The group means are significantly different.")
    else:
        print("Result: Fail to reject the null hypothesis. No significant difference between group means.")


np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)  
group2 = np.random.normal(loc=105, scale=10, size=30)  

independent_t_test(group1, group2)
```

Question 09.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
#Answer

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  
    se = sample_std / np.sqrt(n)       

    
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    
    margin_of_error = t_crit * se

    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
    
    return lower_bound, upper_bound


np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=25)
confidence_interval(sample_data, confidence=0.95)
```

Question10.Write a Python program to calculate the margin of error for a given confidence level using sample data?

```python
#Answer

import math
import statistics
from scipy.stats import norm

def calculate_margin_of_error(data, confidence_level):
    """
    Calculate the margin of error for a given confidence level using sample data.
    
    Parameters:
    data (list): The sample data points.
    confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).

    Returns:
    float: The margin of error.
    """
    if not data:
        raise ValueError("Sample data cannot be empty.")
    
    n = len(data)
    mean = statistics.mean(data)
    std_dev = statistics.stdev(data)
    
    
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    
    margin_of_error = z_score * (std_dev / math.sqrt(n))
    
    return margin_of_error


if __name__ == "__main__":
    sample_data = [50, 52, 47, 49, 51, 53, 48, 50, 52, 49]
    confidence = 0.95  
    
    moe = calculate_margin_of_error(sample_data, confidence)
    print(f"Margin of Error (at {confidence*100:.0f}% confidence): {moe:.2f}")
```

Question11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

```python
#Answer

def bayesian_inference(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):
    """
    Apply Bayes' Theorem to calculate the posterior probability.
    
    Parameters:
    prior_A (float): Prior probability of event A (e.g., having the disease)
    likelihood_B_given_A (float): Likelihood of B given A (e.g., test is positive if disease is present)
    likelihood_B_given_not_A (float): Likelihood of B given not A (e.g., test is positive if no disease)

    Returns:
    float: Posterior probability P(A|B)
    """
    prior_not_A = 1 - prior_A
    
    
    marginal_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    
    
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / marginal_B
    
    return posterior_A_given_B


prior_disease = 0.01                   
sensitivity = 0.99                     
false_positive_rate = 0.05             

posterior = bayesian_inference(prior_disease, sensitivity, false_positive_rate)

print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

Question13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
dat ?a

```python
#Answer

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test from observed data.

    Parameters:
    observed (2D list or numpy array): Contingency table of observed frequencies.

    Returns:
    pandas.DataFrame: Expected frequency table.
    """
    observed = np.array(observed)
    
    
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()
    
    
    expected = np.outer(row_totals, col_totals) / grand_total
    
    return pd.DataFrame(expected, columns=[f'Col{i+1}' for i in range(observed.shape[1])])


observed_data = [[20, 10],
                 [30, 40]]

expected_df = calculate_expected_frequencies(observed_data)

print("Expected Frequencies:")
print(expected_df.round(2))
```

Question16. Implement an F-test using Python to compare the variances of two random samples ?

```python
#Answer

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    sample1, sample2 (list or array): The two independent samples
    alpha (float): Significance level

    Returns:
    dict: Contains F-statistic, p-value, and test conclusion
    """
    n1, n2 = len(sample1), len(sample2)
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)

    
    if var1 > var2:
        F = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1
    else:
        F = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))  

    result = {
        "F-statistic": F,
        "p-value": p_value,
        "Conclusion": "Reject null hypothesis: Variances are different" if p_value < alpha
                      else "Fail to reject null hypothesis: Variances are equal"
    }

    return result


np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

test_result = f_test(sample1, sample2)
for k, v in test_result.items():
    print(f"{k}: {v}")
```

Question19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA ?

```python
#Answer
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check ANOVA assumptions:
    - Normality for each group (Shapiro-Wilk test)
    - Equal variances (Levene's test)
    - Independence: Cannot be tested statistically here (depends on data collection)
    
    Parameters:
        *groups: list of arrays/lists, each representing a group of samples
        alpha: significance level for tests (default 0.05)
        
    Returns:
        dict with results for normality per group, equal variance, and independence note
    """
    results = {"normality": {}, "equal_variance": None, "independence": "Assumed based on study design"}

    
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        results["normality"][f"group_{i}"] = {
            "statistic": stat,
            "p_value": p,
            "normal": p > alpha
        }

    
    stat, p = levene(*groups)
    results["equal_variance"] = {
        "statistic": stat,
        "p_value": p,
        "equal_variance": p > alpha
    }

    return results



if __name__ == "__main__":
    import numpy as np
    
    np.random.seed(0)
    group1 = np.random.normal(50, 10, 30)
    group2 = np.random.normal(52, 12, 30)
    group3 = np.random.normal(49, 9, 30)
    
    assumptions = check_anova_assumptions(group1, group2, group3)
    
    
    print("ANOVA Assumptions Check:")
    for group, res in assumptions["normality"].items():
        print(f"{group} normality: p = {res['p_value']:.4f} -> {'Normal' if res['normal'] else 'Not normal'}")
    
    eq_var = assumptions["equal_variance"]
    print(f"Equal variances (Levene's test): p = {eq_var['p_value']:.4f} -> {'Equal' if eq_var['equal_variance'] else 'Unequal'}")
    print(f"Independence assumption: {assumptions['independence']}")
```

Question25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups ?

```python
#Answer

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_z_test(success_a, size_a, success_b, size_b, alpha=0.05):
    """
    Perform a two-proportion Z-test.
    
    Parameters:
        success_a (int): number of successes in group A
        size_a (int): total observations in group A
        success_b (int): number of successes in group B
        size_b (int): total observations in group B
        alpha (float): significance level (default 0.05)
        
    Returns:
        z_stat (float): computed Z statistic
        p_value (float): two-tailed p-value
        conclusion (str): interpretation of the test
    """
    counts = np.array([success_a, success_b])
    nobs = np.array([size_a, size_b])
    
    z_stat, p_value = proportions_ztest(counts, nobs)
    
    conclusion = ("Reject the null hypothesis: proportions are significantly different."
                  if p_value < alpha else
                  "Fail to reject the null hypothesis: no significant difference in proportions.")
    
    return z_stat, p_value, conclusion


if __name__ == "__main__":
    
    z, p, result = two_proportion_z_test(45, 100, 30, 80)
    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p:.4f}")
    print(f"Result: {result}")
```

Question26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results ?

```python
#Answer

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt
import seaborn as sns

def f_test_variances(sample1, sample2, alpha=0.05):
    """
    Perform F-test to compare variances of two samples.
    
    Null hypothesis H0: variances are equal (sigma1^2 = sigma2^2)
    Alternative hypothesis Ha: variances are different
    
    Parameters:
        sample1, sample2: array-like data samples
        alpha: significance level
        
    Returns:
        f_stat: calculated F-statistic
        p_value: two-tailed p-value for the test
        conclusion: interpretation string
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)
    
    
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1  
    else:
        f_stat = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1
    
    
    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))
    p_value = min(p_value, 1.0)  
    
    conclusion = ("Reject H0: variances are significantly different."
                  if p_value < alpha else
                  "Fail to reject H0: no significant difference in variances.")
    
    return f_stat, p_value, conclusion


np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=50)  
sample2 = np.random.normal(loc=10, scale=3, size=50)  

f_stat, p_value, conclusion = f_test_variances(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")


plt.figure(figsize=(10,6))
sns.kdeplot(sample1, label=f'Sample 1 (var={np.var(sample1, ddof=1):.2f})', fill=True)
sns.kdeplot(sample2, label=f'Sample 2 (var={np.var(sample2, ddof=1):.2f})', fill=True)
plt.title('Kernel Density Estimate of Two Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Answer

import numpy as np
from scipy.stats import chisquare


np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
observed_counts = np.random.multinomial(n=100, pvals=[0.25, 0.25, 0.25, 0.25])

print("Observed counts:", dict(zip(categories, observed_counts)))


expected_proportions = [0.25, 0.25, 0.25, 0.25]
expected_counts = np.array(expected_proportions) * sum(observed_counts)

print("Expected counts:", dict(zip(categories, expected_counts)))


chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

print(f"\nChi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed distribution significantly differs from expected.")
else:
    print("Fail to reject null hypothesis: no significant difference from expected distribution.")
```
